# Hands-on TF.2: Linear Regression in TensorFlow

Cas d'estudi d'una regressi√≥ lineal: y = 0.1 x + 0.3

## Crear dades en un espai bidimensional

Utilitzar el programa per crear els punts dins l'espai

```python
import numpy as np
num_puntos = 100
conjunto_puntos = []
for i in range(num_puntos):
    x1= np.random.normal(0.0, 0.55)
    y1= x1 * 0.1 + 0.3 + np.random.normal(0.0, 0.03)
    conjunto_puntos.append([x1, y1])
x_data = [v[0] for v in conjunto_puntos]
y_data = [v[1] for v in conjunto_puntos]
```

```python
import matplotlib.pyplot as plt
plt.plot(x_data, y_data, 'ro', label='Original data')
plt.legend()
plt.show()
```

## Utilitzar TensorFlow per buscar la l√≠nia que m√©s s'aproxima als punts

Utilitzar l'API de Python pel Machine Learning

```python
import tensorflow as tf
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b
```

Definir la funci√≥ de cost

```python
loss = tf.reduce_mean(tf.square(y - y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)
```

TensorFlow carrega el gr√†fic i executa les operacions en bucle

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
for step in range(8):
    sess.run(train)
    ww = sess.run(W)
    bb = sess.run(b)
    print(step, ww, bb)
    plt.plot(x_data, y_data, 'ro', label='Original data')
    x_data_line = [-1.0,1.5]
    y_data_line = ww*x_data_line+bb
    plt.plot(x_data_line, y_data_line)
    plt.legend()
    plt.show()
```
